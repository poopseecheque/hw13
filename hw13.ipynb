import numpy as np
import pandas as pd
income = np.array([12000, 15000, 13000, 16000, 17000, 15500, 14000, 16500, 15800, 14500])
income_without_tax = income * 0.7

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct']
expenses = np.array([8000, 9000, 9500, 11000, 12000, 11500, 10000, 10800, 10200, 9500])
data = {
    'Month': months,
    'Income_without_tax': income_without_tax,
    'Expenses': expenses
}
df = pd.DataFrame(data)
print("Company Finances:")
print(df)


df['Savings'] = df['Income_without_tax']-df['Expenses']
print(df)


**TASK 1.1**

```python
import matplotlib.pyplot as plt

plt.plot(months, income,'*--b', label="Income")
plt.title("Income by Month")
plt.xlabel("Month")
plt.ylabel("Income ($)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#column chart

savings = df['Savings']

plt.bar(months, savings, color='magenta')
plt.title("Savings by Month")
plt.xlabel("Month")
plt.ylabel("Savings")
plt.grid(True, axis='y')
plt.show()
```

```python
positive_savings = savings[savings >= 0]
positive_months = np.array(months)[savings >= 0]

plt.figure(figsize=(6, 6))
plt.pie(positive_savings, labels=positive_months, autopct='%1.1f%%')
plt.title("Share from total yearly savings")
plt.show()
```

```python
import seaborn as sns
car_crashes = sns.load_dataset("car_crashes")
car_crashes = pd.DataFrame(car_crashes)
car_crashes[:5]
```

```python
sns.lmplot(x="speeding", y="alcohol", data=car_crashes)
plt.title('Speeding vs Alcohol Accidents')
plt.xlabel('% Of Drivers Involved In Fatal Collisions - Speeding')
plt.ylabel('% Of Drivers Involved In Fatal Collisions - Alcohol-Impaired')
plt.show()
#analyses the relationship between speeding-related accidents and alcohol-impaired accident
```

**Based on this chart, the number of speeding accidents is posititvely associated alcohol-impaired accidents**

```python
# Create a bar plot of total accidents by state
car_crashes=car_crashes.sort_values(by='total', ascending = False)
plt.figure(figsize=(10,5))
sns.barplot(x='abbrev', y="total", data=car_crashes)
plt.xticks(rotation=90)
plt.title('Fatal car accidents by state')
plt.xlabel('state')
plt.ylabel('N of accidents')
plt.show()
```

**this graph visually represents the number of crashes, however it is not informative since it doesn't account for the state population, hence we cannot really compare anything here, because we do not have the population variable in the dataset.**

```python
sns.pairplot(car_crashes[['speeding', 'alcohol', 'not_distracted', 'no_previous']])
plt.suptitle("Pairplot of Different Accident Causes", y=1.05)
plt.show()
```

**here there seems to be a positive correlation between the percentage of alcohol-impaired drivers and other factors like speeding and not being distracted. this means states with higher alcohol-impaired drivers also tend to have higher rates of these other factors.
the plots for "no_previous" accidents appear  scattered, suggesting a weaker or less clear relationship between drivers with no previous accidents and the other factors**


```python
correlations = car_crashes.corr()
sns.heatmap(correlations, annot=True, cmap='jet')
plt.title("Correlation matrix of car crashes in the US")
plt.show()
```

**previous conclusions hold here
Both insurance premiums and losses show little correlation with the causes of crashes (speeding, alcohol, etc.), suggesting that insurance factors might be more related to regional economics or legal environments **
